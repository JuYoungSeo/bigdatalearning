```python
def recommendation_user(user_id,k):
    temp = []
    for i in reco_corr.loc[user_id].sort_values(ascending=False)[1:k].index:
        temp.append(set(data[data.user_id==i].item_id) - set(data[data.user_id==user_id].item_id))
    
    print(temp)
    return [data[(data.item_id.isin(temp)) & (data.rating==5) &
                  (data.user_id==i)].sort_values('rating',ascending=False) for
           _ in temp]
    
```

```python
def recommendation_user(user_id,k,n=10):
    s = reco_corr.loc[user_id].sort_values(ascending=False)[1:k+1].index
    temp = pd.DataFrame(columns =['user_id','item_id','rating'])
    for j,i in enumerate(s,start=1):
        t = set(data[data.user_id==i].item_id) - set(data[data.user_id==user_id].item_id)
        # user_id와 유사한 상위 4개개의 id가 i
        # 그 i가 보는 영화 data가 t
        k = data[(data.item_id.isin(t)) & (data.rating==5) &
             (data.user_id==i)]
        k.rating = k.rating*(1-0.05*j)
        # k는 다른 user i가 보는 것중 내가 안보는 것중 rating이 5인 것.
        temp = pd.concat([temp,k])
        # rating이 5이면서 4명을 다 합쳐주면 temp
    else:
           
 #       return  [_ for _ in temp.rename(columns={'item_id':'movie_id'})\
 #   .merge(item[['movie_id','movie_title']])\
 #   .drop_duplicates().groupby('movie_title')\
 #   .mean().sort_values('rating',ascending=False)[:n].index]
           return  [_ for _ in temp.rename(columns={'item_id':'movie_id'})\
    .merge(item[['movie_id','movie_title']])\
    .drop_duplicates().groupby('movie_title')\ 
    .mean().sort_values('rating',ascending=False)[:n].index]
    # drop_duplicates는 완벽히 다 같을때 삭제한다.
```
