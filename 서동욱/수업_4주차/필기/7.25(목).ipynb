```python
# 캐글대회에서 거의다 씀
# 15년전만 해도 통계하는 사람들은 statisic learning으로 알려줬지만
# 이제 기계학습이라고 함.
# 수학, cs과가 했다.
데이터 이해, 프로그래밍, 미적분, 선형대수, 확률통계
인공지능에서 '학습'이 다양하다.
요즘 회사는 구현하는 사람, 알고리즘, 데이터 세분화 되어있다.
보통 학위가 없으면 안시켜주고 데이터 전처리만 시켜준다.
최소 분석하려면 학위가 좀 필요하다. 점점 내려가긴 한다. 그러려면
구현을 잘해야 한다. python이나 R은 느려서, C++이나 java를 코팅하는 작업이 필요하다.
그쪽 사람이 제일 부족하다.
정부는 학부생들이 분석한다고 하면 프로그래밍 포기하는 애들 많아서 포기해서
제2대안으로 하는 것 아니냐는 부정적인 인식을 가지고 있다.
예전 통계학과는 R배우는 사람도 별로 없었는데 이제 C++,java까지 배우는 사람도 많다.

넷플릭스 대회
- 연구하는 사람들은 가난하다.
- netflix가 큰 돈 내는 대회. cs가 이겼는데, 설명 불가능해서 적용 못함.

TED : 세명이 지나가는 도로에서, 박을수 밖에 없는 상황. 자율주행자동차가 어떻게 할까?
=> 분석가능하지 않으면 팔수가 없다.

cs사람들은 결과(예측)에 관심있다. 분석하는 사람은 설명에 관심있다.
결과를 높여야 돈을 벌 수 있다.
자기 도메인에 반대쪽으로 가면 괜찮을 것이다.
NIPS 뉴럴 네트워크 가장 좋은 학회.

Transfer learning 학습된 모델을 조금 변형시키면 다른 것에 쓸 수 있다.
GAN이 2016년도에 나왔는데 이름붙은게 엄청 많다.
Data로부터 Specific 문제 해결을 위한 최적의 모델 만들기
기계학습은 일반화된 알고리즘을 만들 수 없다.
Bias : 결과와 떨어진 정도 높으면 결과와 차이가 크다.
master algorithm : 일반화된 알고리즘 by Pedro domingos 
    
variance와 bias는 trade-off관계여서 한쪽이 커지면 다른 쪽은 작아져야 한다.
오차를 줄여야 하지, 같은 상황에서는 둘다 줄이기 어렵다.

underfit : high bias, low variance
overfit : low bias, high variance
그 중간을 찾길 바람. : model selection
- 알고리즘 바꾸는 것, hyperparameter 바꿔보기.
```
