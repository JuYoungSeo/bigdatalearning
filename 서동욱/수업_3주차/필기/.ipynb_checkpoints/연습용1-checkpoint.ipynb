```python
import requests
import time

def download(method, url, 
             param = None, data = None, 
            timeout = 1,maxretries = 3):
    
    headers = {'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36'}
    #headers = None
    
    try:
        resp = requests.request(method, url,
                                params= param,
                                data=data, headers=headers)
        resp.raise_for_status()
        
    except requests.exceptions.HTTPError as e:
        if 500 <= e.response.status_code < 600 and maxretries > 0: 
            time.sleep(timeout) # param에따라 몇초 기다릴지 결정 할 수가 있다.
            print(maxretries) # 재귀적으로 자기 자신을 부르게 코드를 짜면 된다.
            resp = download(method,
                            url, param = param, data=data,
                            timeout=timeout,
                            maxretries = maxretries - 1)
        else:
            print(e.response.status_code)
            print(e.response.reason)
    return resp
```

```python
def checkBlog(url):
     return compat.urlparse(url)[1] == 'blog.naver.com'
```

```python
def parseURL(seed,depth=3):
    html2 = download('get',seed[0])
    dom2 = BeautifulSoup(html2.text,'lxml')
    
    if len(dom2.select("#mainFrame")) < 1 or seed[1] == depth:
           return []
    
    seed2 = compat.urljoin(seed[0],dom2.select('#mainFrame')[0]['src'])
    html3 = download('get',seed2)
    dom3 = BeautifulSoup(html3.text,'lxml')
           
           
    return [(_["href"],seed[1]+1) for _ in dom3.select('a') 
            if _.has_attr('href') and checkBlog(_["href"])
           and _["href"] not in  ['https://blog.naver.com','https://blog.naver.com/MyBlog.nhn'] ]
 #          and _["href"] != compat.urljoin(url,compat.urlparse(url)[2])
 #           and seed[1] < 3 ]
```

```python
import sqlite3 as sql3
con = sql3.connect("NaverBlog.db")
cur = con.cursor()
```

```python
cur.executescript("""
    DROP TABLE IF EXISTS table1;
    CREATE TABLE table1(
    id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL,
    table2_id INTEGER NOT NULL,
    param TEXT NOT NULL,
    depth INTEGER NOT NULL,
    inbound INTEGER NOT NULL,
    seen BOOLEAN DEFAULT FALSE,
    date TIMESTAMP DEFAULT CURRENT_TIMESTAMP NOT NULL
    );
    
    DROP TABLE IF EXISTS table2;
    CREATE TABLE table2(
    id INTEGER PRIMARY KEY AUTOINCREMENT NOT NULL,
    path TEXT NOT NULL,
    date TIMESTAMP DEFAULT CURRENT_TIMESTAMP NOT NULL
    );

""")
con.commit()
```

```python
baseList = [(_["href"] ,1)
              for _ in dom.select('a.sh_blog_title._sp_each_url._sp_each_title')
                if checkBlog(_["href"])]

netloc = '://'.join(compat.urlparse(baseList[0][0])[:2])

for _ in baseList:
    _urlparse = compat.urlparse(_[0])
    cur.execute("""
        INSERT INTO table2(path) VALUES(?)
    """,[_urlparse[2]])
    con.commit()
    
    cur.execute("""
        SELECT id FROM table2 WHERE path = ? LIMIT 0,1
    """,[_urlparse[2]])
    pathID = cur.fetchone()
    
    cur.execute("""
        INSERT INTO table1(table2_id,param,depth,inbound)
        VALUES(?,?,?,?)
        """,[pathID[0],_urlparse[4],_[1],0])
    con.commit()
```

```python
while True:
    cur.execute("""
    SELECT table1.id, table1.param,
           table1.depth, table2.path, table2.id
    FROM table1
    JOIN table2 ON table1.table2_id = table2.id
    WHERE table1.seen = FALSE and table1.depth < 6
    ORDER BY table1.date ASC
    LIMIT 0,1
    """)
    seed = cur.fetchone()
    if not seed:
        break
    
    cur.execute("""
    UPDATE table1
    SET seen=TRUE
    WHERE id = ?
    """,[seed[0]])
    con.commit()
    
    #print(baseURL)
    baseURL = ("{0}{1}?{2}".format(netloc,seed[3],seed[1]),seed[2])
    
    for href in parseURL(baseURL,5):
        _urlparse = compat.urlparse(href[0])
        cur.execute("""
        SELECT id FROM table2 WHERE path = ? LIMIT 0,1
        """,[_urlparse[2]])
        pathID = cur.fetchone()
        
        if not pathID:
            cur.execute("""
            INSERT INTO table2(path) VALUES(?)
            """,[_urlparse[2]])
            con.commit()
            cur.execute("""
            SELECT id FROM table2 WHERE path = ? LIMIT 0,1
            """,[_urlparse[2]])
            pathID = cur.fetchone()
        
        cur.execute("""
        SELECT id FROM table1 WHERE table2_id = ? and param = ? LIMIT 0,1
        """,[pathID[0],_urlparse[4]])
        netID = cur.fetchone()
        
        if not netID : 
            cur.execute("""
            INSERT INTO table1(table2_id,param,depth,inbound)
            VALUES(?,?,?,?)
            """,[pathID[0],_urlparse[4],seed[2]+1,seed[4]])
            con.commit()
```

```python
queue = list()
queue.extend([(_["href"] ,1)
              for _ in dom.select('a.sh_blog_title._sp_each_url._sp_each_title')
                if checkBlog(_["href"])])

seen = list()
depth = 5

while queue:
    baseURL = queue.pop(0)
    seen.append(baseURL)
    time.sleep(0.1)
    
    linkList = parseURL(baseURL,depth)
    for link in linkList:
        if link not in queue and link not in seen:
            queue.append(link)
    print("Queue : {0}, Seen : {1}".format(len(queue),len(seen)))    
```
