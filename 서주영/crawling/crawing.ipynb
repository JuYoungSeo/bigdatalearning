```python
import requests
import time

def download(method, url, param = None, data = None, timeout=1, maxretries = 3):
    try:
        resq = requests.request(method, url,params = param, data=data)
        resq.raise_for_status()
    except requests.exceptions.HTTPError as e:
        if (500<= e.response.status_code < 600) and maxretries>0:
            time.sleep(timeout)
            print("재시도")
            download(method, url,timeout=timeout,maxretries=maxretries-1)
        else:
            print(e.response.status_code)
            print(e.response.reason)
    return resq

url = 'https://www.crawler-test.com/status_codes/status_500'
headers = {"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/75.0.3770.100 Safari/537.36"}
resq = download("get",url,param={'params':'values'})
print(resq.text)
```
